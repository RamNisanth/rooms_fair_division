```python
def divide_into_two(arr):
    arr1 = []
    arr2 = []
    arr3 = []
    for i in arr:
        if i >= 100 and i <= 199:
            arr1.append(i)
        elif i >= 200 and i <= 299:
            arr2.append(i)
        elif i >= 300 and i <= 399:
            arr3.append(i)
    main = 3
    if len(arr1)>len(arr2):
        if len(arr1)>len(arr3):
            main = 1
    elif len(arr2)>len(arr1):
        if len(arr2)>len(arr3):
            main = 2
        
    m1 = len(arr1)
    m2 = len(arr2)
    m3 = len(arr3)
    Mmax = max(m1,m2,m3)
    if (m1+m2+m3)%2:
        Nmax = (m1+m2+m3)//2+1
    else:
        Nmax = (m1+m2+m3)//2
#     print("Nmax",Nmax)
#     print("mMAX", Mmax)
#     print("main", main)

    if Mmax >= Nmax:
        if(main == 1):
            sorted_list = sorted(arr1)
            Remaining = sorted_list[Nmax:]
            sorted_list = sorted_list[:Nmax]
            Remaining = Remaining + arr2 + arr3

        elif(main == 2):
            sorted_list = sorted(arr2)
            Remaining = sorted_list[Nmax:]
            sorted_list = sorted_list[:Nmax]
            Remaining = Remaining + arr1+arr3
        else:
            sorted_list = sorted(arr3)
            Remaining = sorted_list[Nmax:]
            sorted_list = sorted_list[:Nmax]
            Remaining = Remaining + arr1+arr2

    else:
        concatenated = arr1+arr2+arr3
        sorted_list = concatenated[:Nmax]
        Remaining = concatenated[Nmax:]
    result_set =[sorted_list,Remaining]
    print(result_set)
```
